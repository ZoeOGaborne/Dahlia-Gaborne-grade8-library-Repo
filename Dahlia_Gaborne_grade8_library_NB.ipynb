```python
import firebase_admin
from firebase_admin import credentials, db

if not firebase_admin._apps:
    cred = credentials.Certificate("firebase_key.json")
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://dh-gaborne-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/'
    })
```

```python
import requests

url = "https://raw.githubusercontent.com/ZoeOGaborne/Dahlia-Gaborne-grade8-library-Repo/refs/heads/main/books.json"
response = requests.get(url)
data = response.json()

print(data)
```

```python
ref = db.reference("library")
ref.set(data)
```

```python
ref = db.reference("library")
books_data = ref.get()

print(books_data)
```

```python
print("Book titles:\n")
for book in books_data["books"]:
    print(book["title"])
```

```python
print("Book titles and availability:\n")
for book in books_data["books"]:
    print(book["title"], "-", book["details"]["available"])
```

```python
total = 0
for books in books_data["books"]:
  total += book["details"]["copies"]

print("Total copies of books:", total)
```
